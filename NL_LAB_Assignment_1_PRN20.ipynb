```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import numpy as np

nltk.download('stopwords')
nltk.download('punkt')
```

```python
import pandas as pd
trainData = pd.read_csv("/content/train.csv")
testData = pd.read_csv("/content/test.csv")
```

```python
def lowercase(text):
    text=text.lower()
    return text
```

```python
import string
punc=string.punctuation

def removepunc(text):
    return text.translate(str.maketrans('','',punc))
```

```python
def remove_sw(text):
    new_list=[]
    words=word_tokenize(text)
    sws=stopwords.words('english')
    for word in words:
        if word not in sws:
            new_list.append(word)

    return ' '.join(new_list)
```

```python
def stem_text(text):
    new_list=[]
    stemmer=PorterStemmer()
    words=word_tokenize(text)
    for word in words:
        new_list.append(stemmer.stem(word))

    return ' '.join(new_list)
```

```python
trainData['Description']=trainData['Description'].apply(lowercase)
trainData['Description']=trainData['Description'].apply(removepunc)
trainData['Description']=trainData['Description'].apply(remove_sw)
trainData['Description']=trainData['Description'].apply(stem_text)

testData['Description']=testData['Description'].apply(lowercase)
testData['Description']=testData['Description'].apply(removepunc)
testData['Description']=testData['Description'].apply(remove_sw)
testData['Description']=testData['Description'].apply(stem_text)
```

```python
trainData.head()
```

```python
testData.head()
```

```python
X_train = trainData['Description']
Y_train = trainData['Class Index']
X_test = testData['Description']
Y_test = testData['Class Index']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
def evaluate_model(model, X_train_vec, X_test_vec, y_train, y_test):
    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_test_vec)
    from sklearn.metrics import accuracy_score, f1_score
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")
    return f"\tAccuracy: {accuracy:.4f}, F1-score (weighted): {f1:.4f}"
```

#### TfidVectorizer

```python
import warnings

from sklearn.feature_extraction.text import TfidfVectorizer
# Vectorize training and testing data
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

models = {
    "LogisticRegression": LogisticRegression(solver="lbfgs"),
    "SVC": SVC(kernel="linear", max_iter = 1000),  # Change kernel based on needs
    "RandomForestClassifier": RandomForestClassifier(n_estimators=100),
}

for model_name, model in models.items():
        print(f"\t- {model_name}:")
        print(evaluate_model(model, X_train_vec, X_test_vec, Y_train, Y_test))
        print("\n\n")
```

#### CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer instance
vectorizer = CountVectorizer()

# Vectorize training data
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

models = {
    "LogisticRegression": LogisticRegression(solver="lbfgs"),
    "SVC": SVC(kernel="linear", max_iter = 1000,),  # Change kernel based on needs
    "RandomForestClassifier": RandomForestClassifier(n_estimators=100),
}

for model_name, model in models.items():
        print(f"\t- {model_name}:")
        print(evaluate_model(model, X_train_vec, X_test_vec, Y_train, Y_test))
        print("\n\n")
```

```python
from gensim.models import KeyedVectors

word2vec_model = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/NLP-GoogleVectors/GoogleNews-vectors-negative300.bin', binary=True)

# Define functions for text preprocessing and word embedding
def preprocess_text(text):
    # Your preprocessing steps here (e.g., cleaning, lowering, removing stopwords)
    words = text.split()
    return words

def word2vec_mean(text):
    words = preprocess_text(text)
    embedding_dim = 300  # Assuming GoogleNews has 300 dimensions
    embeddings = np.zeros(embedding_dim)
    count = 0
    for word in words:
        if word in word2vec_model:
            count += 1
            embeddings += word2vec_model[word]
    if count > 0:
        return embeddings / count
    else:
        return np.zeros(embedding_dim)  # Handle words not in Word2Vec

# Extract word embeddings for each document
embeddings_train = [word2vec_mean(text) for text in (X_train)]
embeddings_test = [word2vec_mean(text) for text in (X_test)]


models = {
    "LogisticRegression": LogisticRegression(solver="lbfgs"),
    "SVC": SVC(kernel="linear", max_iter = 1000,),  # Change kernel based on needs
    "RandomForestClassifier": RandomForestClassifier(n_estimators=100),
}

for model_name, model in models.items():
        print(f"\t- {model_name}:")
        print(evaluate_model(model, embeddings_train, embeddings_test, Y_train, Y_test))
        print("\n\n")
```
